```python
name_array = np.array(['John', 'Mike', 'Sarah', 'Kate', 'Samuel'])
score_array= np.array([78, 95, 84, 98, 88])
```

```python
sort_score=np.argsort(score_array)
print("성적 정렬시 score의 인덱스",sort_score)
print('name_arrary의 이름 출력',name_array[sort_score])
```
