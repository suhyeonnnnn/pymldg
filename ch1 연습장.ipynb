```python
#nan 여부 확인 : isna()
#nan 다른 값 대체: fillna()
titanic_df.isna().sum()
```

```python
titanic_df['Cabin']= titanic_df['Cabin'].fillna('C000') #반환값을 다시 받거나, inplace=True로 해야지  실제 데이터값이 변경됨
titanic_df.head() 
```

```python
#age 칼럼의 결손데이터를 평균값으로 대체, Embarked 칼럼의 결손데이터를 S로 대체
titanic_df['Age']=titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Embarked']=titanic_df['Embarked'].fillna('S')
titanic_df.head()
```

```python
titanic_df.isna().sum()
```

```python
#lambda 식 가공
#name 칼럼의 문자열 개수를 Name_len 칼럼에 만들기
titanic_df['Name_len']= titanic_df['Name'].apply(lambda x: len(x))
titanic_df[['Name','Name_len']].head()
```

```python
#나이가 15세 미만이면 child, 이상이면 adult로 구분하라
def child_adult(x):
    if x < 15:
        return 'Child'
    else:
        return 'Adult'

titanic_df['Child_Adult']=titanic_df['Age'].apply(lambda x: child_adult(x))
titanic_df[['Age','Child_Adult']].head(10)
```

```python
titanic_df['Child_Adult']=titanic_df['Age'].apply(lambda x: 'Child' if x <=15 else 'Adult')
titanic_df[['Age','Child_Adult']].head(10)
```
